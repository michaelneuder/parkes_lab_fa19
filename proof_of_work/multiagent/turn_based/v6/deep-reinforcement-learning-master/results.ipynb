```python
import matplotlib.gridspec as gridspec
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
```

```python
img_path = '../img/'
```

```python
plt.figure(figsize =(48,32))
gs1 = gridspec.GridSpec(2, 3)
gs1.update(wspace=0.03, hspace=0.03)

ax_dict = {}
for ii in range(6):
    ax_dict[ii] = plt.subplot(gs1[ii])
    ax_dict[ii].set_xticklabels([])
    ax_dict[ii].set_yticklabels([])
    ax_dict[ii].get_xaxis().set_visible(False)
    ax_dict[ii].get_yaxis().set_visible(False)
    
pic_array = []
for i in range(1,7):
    pic_array.append(mpimg.imread(os.path.join(img_path, '45_{}.png'.format(i))))

for i in range(6):
    ax_dict[i].imshow(pic_array[i])
plt.savefig('45_progression.png')
plt.show()
```
