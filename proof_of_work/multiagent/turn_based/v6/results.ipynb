```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.style.use('seaborn-whitegrid')
```

```python
ell = int(1e6)
alpha = 0.4
```

```python
x_1 = np.arange(0, 1.1, 0.1)
opt = [0.247946, 0.217535, 0.185891, 0.158213, 0.129615, 
       0.099756, 0.079739, 0.060014, 0.039963, 0.020398, 0.000000]
honest = [0.200213, 0.180435, 0.160371, 0.140537, 0.119859, 
          0.100504, 0.080226, 0.060702, 0.039587, 0.020180, 0.000631]
sm1 = [0.188279, 0.161661, 0.132938, 0.107686, 0.081527, 0.054360,
       0.026929, 0.000997, -0.025712, -0.051792, -0.078537]
```

```python
plt.style.use('fivethirtyeight')
f, ax = plt.subplots(figsize=(10,7))
ax.plot(x_1, opt, 'r--', alpha=0.5, label='optimal')
ax.plot(x_1, sm1, 'purple', linestyle='dashdot', alpha=0.5, label='sm1')
x = np.arange(0, 1, 0.001)
ax.plot(x_1, honest, 'ko', alpha=1, label='honest (simulated)')
ax.plot(x, alpha*(1-x)/2, 'g', linestyle='dotted', alpha=0.5, label='honest (theoretical)')
ax.set_xticks(x_1)
ax.set_ylabel(r'$\rho-$reward per action')
ax.set_xlabel(r'$C-$cost of mining')
ax.set_title(r'$\alpha=0.4$')
plt.legend(frameon=True, framealpha=1)
f.set_facecolor('white')
plt.show()
```
