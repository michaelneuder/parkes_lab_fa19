```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ell = int(1e6)
alpha = 0.4
```

```python
x_1 = np.arange(0, 1.1, 0.1)
opt = [0.281887, 0.24953023999804888, 0.21765700000341595, 0.18418044000208025, 
       0.15210635999789934, 0.12172000000154946, 0.09110967999930931, 0.06349264000055568, 
       0.040268000000126654, 0.01986299999985598, 0.0]
honest = [0.200142, 0.17971599999993004, 0.15982600000053152, 0.14011700000097527, 
          0.12029699999875021, 0.10078400000090008, 0.08021299999979413, 0.05974900000038549, 
          0.03920100000012729, 0.019849999999861163, -3.9000000001506397e-05]
sm1 = [0.253527, 0.22576483999877398, 0.19856788000229686, 0.17036588000167652, 
       0.1436748399981674, 0.11619000000133672, 0.08794767999943354, 0.06059072000053416, 
       0.034018240000168905, 0.005980160000051717, -0.022351000000146837]
```

```python
plt.style.use('fivethirtyeight')
f, ax = plt.subplots(figsize=(10,7))

ax.plot(x_1, opt, 'r--', alpha=0.5, label='optimal')
ax.plot(x_1, sm1, 'purple', linestyle='dashdot', alpha=0.5, label='sm1')

x = np.arange(0, 1, 0.001)
ax.plot(x_1, honest, 'ko', alpha=1, label='honest (simulated)')
ax.plot(x, alpha*(1-x)/2, 'g', linestyle='dotted', alpha=0.5, label='honest (theoretical)')

ax.set_xticks(x_1)
ax.set_ylabel(r'$\rho-$reward per action')
ax.set_xlabel(r'$C-$cost of mining')
ax.set_title(r'$\alpha=0.4$')
plt.legend(frameon=True, framealpha=1)
f.set_facecolor('white')
# ax.set_xlim([0.6,0.9])
# ax.set_ylim(0.0, 0.1)
plt.show()
```
