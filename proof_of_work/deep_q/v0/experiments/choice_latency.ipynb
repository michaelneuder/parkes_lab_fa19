```python
import matplotlib.pyplot as plt
import numpy as np
import progressbar 
import time
```

```python
choice_results = []
bar = progressbar.ProgressBar()
for n in bar(range(3000)):
    time_1 = time.time()
    for _ in range(n):
        np.random.choice([0,1,2])
    time_2 = time.time()
    choice_results.append(time_2 - time_1)
```

```python
randint_results = []
bar = progressbar.ProgressBar()
for n in bar(range(3000)):
    time_1 = time.time()
    for _ in range(n):
        np.random.randint(low=0, high=3)
    time_2 = time.time()
    randint_results.append(time_2 - time_1)
```

```python
plt.style.use('ggplot')
f, ax = plt.subplots(figsize=(10,10))
ax.set_ylabel('latency (sec)')
ax.set_xlabel('number of random selections')
ax.plot(choice_results, label='choice')
ax.plot(randint_results, label='randint')
plt.legend()
plt.savefig('choice_latency.png')
plt.show()
```
