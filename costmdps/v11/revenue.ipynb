```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import numpy as np
```

```python
x = np.arange(4032)
x_time = x
y_honest = .400 *x
x_half = np.arange(2016)
x_half_time = x_half
y_selfish_half = .3261*x_half
x_half_2 = np.arange(2016, 4032)
x_half_2_time = x_half_2
y_selfish_half_2 = .41666 *x_half_2 -182.4971
```

```python
f, ax = plt.subplots(figsize=(8,5))
f.set_facecolor('white')
ax.set_facecolor('white')
ax.plot(x_time, y_honest, color='orange', label='honest')
ax.set_xlabel(r'$t-$ time (minutes)')
ax.set_ylabel(r'$\rho-$ cumulative revenue')
ax.plot(x_half_time, y_selfish_half, color='red', label=r'selfish $(t<2016)$')
ax.plot(x_half_2_time, y_selfish_half_2, color='green', label=r'selfish $(t>2016)$')
plt.legend(facecolor='white')
plt.show()
```

```python
x = np.arange(2016*6)
y_honest = .400 *x
x_pre = np.arange(2016)
y_selfish_pre = .3261*x_pre
x_post = np.arange(2016, 2016*6)
y_selfish_post = .41666 *x_post -182.4971
```
