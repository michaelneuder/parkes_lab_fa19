```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ell = int(1e6)
alpha = 0.4
```

```python
x_1 = np.arange(0, 1.1, 0.1)
opt = [0.252455, 0.2222682799987051, 0.19055108000200116, 0.1641019600015836, 
       0.1359517199982273, 0.11079120000124187, 0.08377139999959356, 
       0.059025000000377525, 0.03971300000012723, 0.019824999999852943, 0.0]
honest = [0.199588, 0.18026099999992684, 0.15973800000052185, 0.14003500000097432, 
          0.12044299999874539, 0.09957000000089293, 0.0795879999997974, 0.05988200000038245, 
          0.039919000000126714, 0.019467999999861065, 0.0002620000000019822]
sm1 = [0.235345, 0.2112663599991296, 0.18433584000169267, 0.15931704000145552, 
       0.13416291999832505, 0.10783120000119863, 0.0828729599996225, 0.05670960000047274, 
       0.029758360000154017, 0.004633160000016437, -0.020216800000034302]
```

```python
plt.style.use('fivethirtyeight')
f, ax = plt.subplots(figsize=(10,7))

ax.plot(x_1, opt, 'r--', alpha=0.5, label='optimal (simulated)')

ax.plot(x_1, sm1, 'purple', linestyle='dashdot', alpha=0.5, label='sm1')

x = np.arange(0, 1, 0.001)
ax.plot(x_1, honest, 'ko', alpha=0.5, label='honest (simulated)')
ax.plot(x, alpha*(1-x)/2, 'g', linestyle='dotted', alpha=0.5, label='honest (theoretical)')

ax.set_xticks(x_1)
ax.set_ylabel(r'$\rho-$reward per action')
ax.set_xlabel(r'$C-$cost of mining')
ax.set_title(r'$\alpha=0.4$')
plt.legend(frameon=True, framealpha=1)
f.set_facecolor('white')
ax.set_xlim([0.5,0.9])
ax.set_ylim(0.0, 0.1)
plt.show()
```
