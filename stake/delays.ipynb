```python
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('fivethirtyeight')
mpl.rcParams['axes.facecolor'] = 'w'
mpl.rcParams['figure.facecolor'] = 'w'
mpl.rcParams['legend.framealpha'] = 1
import numpy as np
import scipy.special as ss
```

 \mathcal{D}_2 &= 8 \cdot (5  (p_\ell - 2) + \max(24-e_\ell, 0) - \max(e_\ell-8, 0))

```python
e = np.arange(0,33)
```

```python
results = []
for x in range(1,4):
    xs = []
    for _e in e:
        xs.append(8 * (5 * (x - 2) + max(24-_e, 0) - max(_e-8, 0)))
    results.append(xs)
```

```python
f, ax = plt.subplots(figsize=(7,5))
ax.plot(e, results[0], 'rx', label=r'$p_{\ell}=1$')
ax.plot(e, results[1], 'o', color='purple', label=r'$p_{\ell}=2$')
ax.plot(e, results[2], 'g+', label=r'$p_{\ell}=3$')
ax.set_xlabel(r'$e_{\ell}$')
ax.set_ylabel(r'$D_2$')
plt.legend()
plt.show()
```

\mathcal{D}_2 &= 8 \cdot (5\cdot (p_\ell - n_{\ell+1}) +12)

```python
p = np.arange(1,15)
results = []
for n in range(2,5):
    x_1 = []
    for _p in p:
        x_1.append(8 * (5 * (_p - n) + 4) )
    results.append(x_1)
```

```python
f, ax = plt.subplots(figsize=(7,5))
ax.plot(p, results[0], 'rx', label=r'$n_{\ell}=2$')
ax.plot(p, results[1], 'o', color='purple', label=r'$n_{\ell}=3$')
ax.plot(p, results[2], 'g+', label=r'$n_{\ell}=4$')
ax.set_xlabel(r'$p_{\ell}$')
ax.set_ylabel(r'$D_2$')
plt.legend()
plt.show()
```

\mathcal{D}_2 &= 8 \cdot (5  (1 - n_{\ell+1}) + \max(24-e_\ell, 0) - \max(e_\ell-8, 0))

```python
e = np.arange(0,33)
results = []
for n in range(2,5):
    x_1 = []
    for _e in e:
        x_1.append(8 * (5 * (1 - n) + max(24-_e, 0) - max(_e-8, 0)))
    results.append(x_1)
```

```python
f, ax = plt.subplots(figsize=(7,5))
ax.plot(e, results[0], 'rx', label=r'$n_{\ell+1}=2$')
ax.plot(e, results[1], 'o', color='purple', label=r'$n_{\ell+1}=3$')
ax.plot(e, results[2], 'g+', label=r'$n_{\ell+1}=4$')
ax.set_xlabel(r'$e_{\ell}$')
ax.set_ylabel(r'$D_2$')
plt.legend()
plt.show()
```
